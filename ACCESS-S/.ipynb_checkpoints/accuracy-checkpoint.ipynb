# Functions

```python

# All of the following need two functions, the second is to apply the second groupy (groupy year)

def group_count(group):
    return group.groupby('time.year').count()


def percent_rain(data):
    rain_year = data.groupby('time.month').apply(group_count)
    days_year = data.fillna(0).groupby('time.month').apply(group_count)
    
    return rain_year/days_year








def group_count_main(data):
    rain_year = data.groupby('time.month').apply(group_count)

    
    return rain_year








def group_resample_mean_2(group):
        return group.groupby('time.year').mean()

def group_resample_mean(group):
        return group.groupby('time.month').apply(group_resample_mean_2)
    
    
    
    

def group_resample_sum_2(group):
        return group.groupby('time.year').sum()

def group_resample_sum(group):
        return group.groupby('time.month').apply(group_resample_sum_2)

```

```python
def calculate_score(data, climatology, _type_ = 'greater'):
    
    phases = data.phase.values
    months = data.month.values
    
    phase_stor = []
    
    for phase in phases:
        month_stor = []
        for month in months:


            sub_set = data.sel(phase = phase, month = month)

            # This gets rid of all the years with no data
            sub_set = sub_set.where(sub_set.precip > 0, drop = True)
            
            # Number of years where there is data
            max_score = len(sub_set.year.values)

            climatology_subset = climatology.sel(phase = phase, month = month)
            
            if _type_ == 'greater':
                score = sub_set.where(sub_set.precip > climatology_subset.precip)
            else:
                score = sub_set.where(sub_set.precip < climatology_subset.precip)


            score = score.where(~np.isfinite(score.precip),1)
            
            month_stor.append(score)

        xr_phase = xr.concat(month_stor, pd.Index(months, name = 'month'))

        phase_stor.append(xr_phase)

    full_score = xr.concat(phase_stor, dim = 'phase')
    
    return full_score
```

```python
# This calculates how many samples from each of the months and phases have actually been used

def calculate_functional_year(awap):
    
    phases = awap.phase.values
    months = awap.month.values
    
    phase_stor = []

    for phase in phases:
        month_stor = []
        for month in months:


            sub_set = awap.sel(phase = phase, month = month)

            # This gets rid of all the years with no data
            sub_set = sub_set.where(sub_set.precip > 0, drop = True)
            
            # Number of years where there is data
            max_score = len(sub_set.year.values)
            
            month_stor.append(max_score)


        phase_stor.append(month_stor)
        
    functional_years = xr.Dataset({'max_score':(('phase','month'), np.array(phase_stor))},
                                {
                                    'phase':phases,
                                    'month':months
                                })
    
    return functional_years 
```

```python
#!!!!!!! UN-USED !!!!!!!
def calculate_score_access(data, awap, climatology, _type_ = 'greater'):
    
    # This fucntion calculates the score for each year, whether it is above or below the climatology.
    
    
    phase_stor = []

    for phase in phases:
        month_stor = []
        for month in months:


            sub_set = data.sel(phase = phase, month = month)
            awap_sub = awap.sel(phase = phase, month = month)

            # This gets rid of all the years with no data
            sub_set = sub_set.where(sub_set.precip > 0, drop = True)
            awap_sub = awap_sub.where(awap_sub.precip > 0, drop = True)

            
            # Number of years where there is data:
                 # the number of year where data actually occurs in awap x the number of different ensemble member
            max_score = len(awap_sub.year.values) * 11 

            climatology_subset = climatology.sel(phase = phase, month = month)
            
            if _type_ == 'greater':
                score = sub_set.where(sub_set.precip > climatology_subset.precip)
            else:
                score = sub_set.where(sub_set.precip < climatology_subset.precip)


            score = score.where(~np.isfinite(score.precip),1)
            
            month_stor.append(score/max_score)

        xr_phase = xr.concat(month_stor, pd.Index(months, name = 'month'))

        phase_stor.append(xr_phase)

    full_score = xr.concat(phase_stor, dim = 'phase')
    
    return full_score
```

```python
def accuarcy_access(awap_func_score_neg, awap_func_score_pos, functional_years, function, 
                   climatology):
    ensembles = access_split.ensemble.values
    ens_stor = []

    for ensemble in ensembles:


        # Getting a single ensemble
        single_access = access_split.sel(ensemble = ensemble)

        # Groupby month and resample year, and then apply the function
        access_year_func  = single_access.apply(function)

        # Find all the instances when the ensemble is greater and less than the cliatology
        access_func_score_pos = calculate_score(access_year_func, climatology )
        access_func_score_neg = calculate_score(access_year_func, climatology,
                                                        _type_ = 'less' )

        # Make sure that all the times when the ensemble is greater (or less) than the climatology
        # that awap was also greater than the climatolgy at this instance
        access_func_score_pos = access_func_score_pos.where(\
                                    (access_func_score_pos == 1) &  (awap_func_score_pos == 1)).fillna(0)
        access_func_score_neg = access_func_score_neg.where(\
                                    (access_func_score_neg == 1) &  (awap_func_score_neg == 1)).fillna(0)
        
        # Summing together the times when access and awap where both greater and less than 
        # the climatology
        ensemble_correct = (access_func_score_pos + access_func_score_neg).sum(dim = 'year')


        ens_stor.append(ensemble_correct)


    total_correct = xr.concat(ens_stor, pd.Index(ensembles, name = 'ensemble'))
    
    
    # The accuracy is going to the the sum of all the correct scores * 100/
    # the number of ensembles (11) * the number of years that actually hade rainfall data (
    # some months in some years there isn't rainfall data for a given MJO phase)
    accuracy = (total_correct.sum(dim = 'ensemble').precip * 100/\
                (11 * functional_years.max_score)).to_dataset(name = 'precip')
    
    
    # Applying the andrew mask and also the ERA-5 landsea mask
    accuracy = misc.apply_masks(accuracy)
    
    return accuracy

```
