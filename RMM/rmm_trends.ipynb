# Trend Lines

```python
'''Counts the number of days in each of the MJO phases for each wet-season. This is useful for 
normalising all of the count trends'''
def count_in_rmm_phase(rmm):
    
    # We are looking at wet season, so will need to be made into wet season data set
    rmm = phase_calc.wet_season_year(rmm)

    
    rmm_act = rmm.where(rmm.amplitude > 1, drop = True)
    
    phases = np.arange(1,9)
    single_phase = []
    for phase in phases:

         # Just the data for this single rmm phase
        rmm_single_phase = rmm_act.where(rmm_act.phase == phase, drop = True)
         # Resmapling via year, to get the number of days in each phase
        number_per_year = rmm_single_phase.phase.resample(time = 'y').count(dim = 'time')
        # Appending
        single_phase.append(number_per_year.values)



    '''Inactive Phase'''
    rmm_inact = rmm.where(rmm.amplitude <=1 , drop = True)
    number_per_year_inact = rmm_inact.phase.resample(time = 'y').count(dim = 'time')

    single_phase.append(number_per_year_inact.values)

    titles = np.append(np.array([str(phase) for phase in phases]),['inactive'])

    datafile_RMM_split = xr.Dataset({'number':(('phase','year'), single_phase)},
                                   {'phase':titles,
                                    'year': number_per_year.time.dt.year.values
                                   })
    
    
#     datafile_RMM_split = xr.concat(single_phase, pd.Index(titles, name = 'phase'))
    
    return datafile_RMM_split
```

```python
phase_count = count_in_rmm_phase(rmm)
```

```python
# This is the total number of years available
length = len(phase_count.year.values)
length
```

```python
# Chopping of the end year, as it only contains half a wet season
phase_count = phase_count.isel(year = slice(0, length - 1))

phase_count
```

```python
phase_count_percent = phase_count * 100/phase_count.sum(dim = 'phase')
phase_count_percent
```

```python
fig = plt.figure(figsize = (10,20))
gs = gridspec.GridSpec(9,1)
data = phase_count


import mystats
for i,phase in enumerate(data.phase.values):
    ax = fig.add_subplot(gs[i])
    num_days = data.sel(phase = phase).number
    years = num_days.year.values
    
    ax.plot(years,num_days)
    
    
    x = years
    y = num_days.values
    
    idx = np.isfinite(x) & np.isfinite(y)
        
    pval = mystats.mann_kendall(y[idx], return_all = 0)
    
    grad, intc = np.polyfit(x[idx],y[idx],1)
    y  = grad * years + intc
    
    if pval  < 0.05:
        sig = 'p < 0.05'
        plt.plot(years, y, color = 'r')
    else:
        sig = 'Not Significant'
        plt.plot(years, y, color = 'grey')
    
    
    trend_percent = grad * 1000 / np.mean(y)
    

  
    

    
    
    if phase != 'inactive':
        ax.spines['bottom'].set_visible(False)
        ax.set_xticks([])
        ax.set_ylim((0,40))
        

  
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    if phase == 'inactive':
        ax.set_ylabel(phase.capitalize(), rotation = 0, fontsize = 12, labelpad = 20)
    else:
        ax.set_ylabel(phase, rotation = 0, fontsize = 12, labelpad = 20)
    if pval < 0.05:
        ax.annotate(f'+{np.round(trend_percent,1)}% Per Decade\n({str(np.round(grad,1) * 10)} days/decade)', 
                    xy = (0.7,0.82), xycoords = 'axes fraction')
#     ax.legend(bbox_to_anchor = (1,0.6))


ax.annotate('MJO Phase', xy =(0.028, 0.5), xycoords = 'figure fraction', rotation = 90, size = 18, va = 'center')
fig.savefig('rmm_trends/days_in_phase.png', dpi = 500)
```
