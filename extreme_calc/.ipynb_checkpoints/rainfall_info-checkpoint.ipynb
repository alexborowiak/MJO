# Saving Info For All Percentile

```python
def plot_every_month(data, plot_title, cbar_title,  subphase_plot,savedir = savedir):
    
    for month in [10,11,12,1,2,3]:
    
        data_month = data.precip.sel(month = month)
        
        sub_plot_title = f'{calendar.month_name[month]}' + plot_title
        subphase_plot = reload(subphase_plot) 
        subphase_plot.raw_plot(data_month, dont_plot = 1,savefig = 1,
                               cbar_title = cbar_title, plot_title = sub_plot_title, savedir = savedir)
        print(str(month))
```

```python
def generate_all_plots(raw, split, q, savedir, subphase_calc, subphase_plot):
    
    '''This is creating a dataset of events above a certain percentile'''
    above_q = subphase_calc.find_events_above_q(raw, split, q)
    
    '''Calculating the Indices we want to plot'''
    above_q_count = above_q.groupby('time.month').count(dim = 'time')
    above_q_sum = above_q.groupby('time.month').sum(dim = 'time')
    above_q_mean = above_q.groupby('time.month').mean(dim = 'time')
    
    '''Plotting and Saving for every month in the Dataset'''
    print('count')
    cbar_title = 'Number of Events'
    plot_title = f' Number of Rainfall Events > {q}th Percentile \nfor Phases of the MJO'
    plot_every_month(above_q_count,plot_title, cbar_title, subphase_plot, savedir = savedir + 'count/' )
    
    
    print('------------------------- \n sum')
    cbar_title = 'Rainfall (mm)'
    plot_title = f' Total Rainfall > {q}th Percentile \nfor Phases of the MJO'
    plot_every_month(above_q_sum, plot_title, cbar_title, subphase_plot, savedir = savedir +  'sum/')
    
    
    print('------------------------- \n intensity')
    cbar_title = 'Intensity (mm)'
    plot_title = f' Mean Intsity for Rainfall > {q}th Percentile \nfor Phases of the MJO'
    plot_every_month(above_q_mean, plot_title, cbar_title, subphase_plot, savedir = savedir + 'int/')
```

```python
raw = awap.precip
split = awap_split.precip
q = 90

subphase_calc = reload(subphase_calc) 


generate_all_plots(raw, split, q, savedir, subphase_calc, subphase_plot)
```

```python
raw = awap.precip
split = awap_split.precip
q = 95

subphase_calc = reload(subphase_calc) 


generate_all_plots(raw, split, q, savedir, subphase_calc, subphase_plot)
```

```python
raw = awap.precip
split = awap_split.precip
q = 99

subphase_calc = reload(subphase_calc) 


generate_all_plots(raw, split, q, savedir, subphase_calc, subphase_plot)
```
